```python
#Clean specific datanames related
# Limpiar los nombres de las columnas
dataframe_total.columns = dataframe_total.columns.str.strip().str.replace('\n', ' ').str.replace('\r', ' ').str.replace('.', '')
dataframe_total.columns = dataframe_total.columns.str.replace(' +', ' ', regex=True)
print(dataframe_total.columns)
```

```python
#Treatment of the specific nule values +
# Eliminar columnas con más del 50% de valores nulos
threshold = len(dataframe_total) * 0.50
dataframe_total.dropna(thresh=threshold, axis=1, inplace=True)

# Imputar valores nulos para columnas numéricas con la mediana y categóricas con la moda
for col in dataframe_total.select_dtypes(include=['number']).columns:
    median = dataframe_total[col].median()
    dataframe_total[col].fillna(median, inplace=True)

for col in dataframe_total.select_dtypes(include=['object']).columns:
    mode = dataframe_total[col].mode()[0]
    dataframe_total[col].fillna(mode, inplace=True)

```

```python
#Convert the columns into the specific type of data based on the appropiated set
# Función para convertir valores con 'K', 'M', etc., a numéricos
def convert_to_numeric(x):
    if isinstance(x, str):
        if 'M' in x:
            return float(x.replace('M', '')) * 1000000
        elif 'K' in x:
            return float(x.replace('K', '')) * 1000
        else:
            return float(x)
    return x

# Aplicar la conversión en las columnas relevantes
cols_to_convert = ['Subscribers', 'avg views', 'avg likes', 'avg comments', 'Followers', 'views(avg)', 'likes(avg)', 'comments(avg)', 'shares(avg)']
for col in cols_to_convert:
    if col in dataframe_total.columns:
        dataframe_total[col] = dataframe_total[col].apply(convert_to_numeric)
```
